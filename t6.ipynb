```python
from collections import deque
```

```python
class Graph:

    def __init__(self, quantidade_vertices):
        self.V = quantidade_vertices
        self.E = 0
        self.lista = []

        for i in range(self.V):
            self.lista.append([])

    def add_edge(self, v, w):
        self.lista[v].append(w)
        self.lista[w].append(v)
        self.E += 1

    def adj(self, v):
        return self.lista[v]

    def mostrar(self):
        for i in range(self.V):
            print(str(i) + ": " + str(self.lista[i]))
```

```python
def ler_arquivo(nome_arquivo):

    with open(nome_arquivo, "r") as arquivo:
        linhas = arquivo.readlines()

    linhas = [linha.strip() for linha in linhas if linha.strip() != ""]

    quantidade_vertices = int(linhas[0])
    quantidade_arestas = int(linhas[1])

    grafo = Graph(quantidade_vertices)

    for i in range(2, 2 + quantidade_arestas):

        v, w = map(int, linhas[i].split())

        if v < 0 or v >= quantidade_vertices:
            raise Exception("Vértice inválido")

        if w < 0 or w >= quantidade_vertices:
            raise Exception("Vértice inválido")

        if v == w:
            raise Exception("Laço não permitido")

        grafo.add_edge(v, w)

    return grafo
```

```python
def bfs(grafo, inicio):

    visitado = [False] * grafo.V
    fila = deque()

    fila.append(inicio)
    visitado[inicio] = True

    while fila:

        atual = fila.popleft()

        for vizinho in grafo.adj(atual):

            if visitado[vizinho] == False:
                visitado[vizinho] = True
                fila.append(vizinho)

    return visitado
```

```python
def eh_arvore(grafo):

    if grafo.E != grafo.V - 1:
        return False, "Número de arestas diferente de V-1"

    visitado = bfs(grafo, 0)

    for item in visitado:
        if item == False:
            return False, "Grafo desconexo"

    return True, "Árvore válida"
```

```python
def encontrar_centros(grafo):

    grau = [0] * grafo.V
    folhas = []

    for i in range(grafo.V):

        grau[i] = len(grafo.adj(i))

        if grau[i] <= 1:
            folhas.append(i)

    processados = len(folhas)

    while processados < grafo.V:

        novas_folhas = []

        for folha in folhas:

            for vizinho in grafo.adj(folha):

                grau[vizinho] -= 1

                if grau[vizinho] == 1:
                    novas_folhas.append(vizinho)

        processados += len(novas_folhas)
        folhas = novas_folhas

    return folhas
```

```python
def montar_arvore(grafo, atual, pai):

    filhos = []

    for vizinho in grafo.adj(atual):

        if vizinho != pai:
            filhos.append(montar_arvore(grafo, vizinho, atual))

    return filhos
```

```python
def codificar(no):

    if len(no) == 0:
        return "()"

    codigos = []

    for filho in no:
        codigos.append(codificar(filho))

    codigos.sort()

    texto = ""

    for codigo in codigos:
        texto += codigo

    return "(" + texto + ")"


def codigo_canonico(grafo):

    centros = encontrar_centros(grafo)

    codigos = []

    for centro in centros:

        raiz = montar_arvore(grafo, centro, -1)
        codigos.append(codificar(raiz))

    codigos.sort()

    return codigos[0]
```

```python
def analisar(nome_arquivo):

    print("=" * 60)
    print("Arquivo:", nome_arquivo)

    try:
        grafo = ler_arquivo(nome_arquivo)

    except Exception as erro:
        print("Erro:", erro)
        return None

    print("\nLista de Adjacência:")
    grafo.mostrar()

    valido, mensagem = eh_arvore(grafo)

    print("\nValidação:", mensagem)

    if valido == False:
        return None

    centros = encontrar_centros(grafo)
    print("Centro(s):", centros)

    codigo = codigo_canonico(grafo)
    print("Código Canônico:", codigo)

    return grafo
```
